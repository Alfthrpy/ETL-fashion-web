```python
import pandas as pd

df = pd.read_csv('testing.csv')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
dirty_pattern = {
    "Title" : ['Unknown Product'],
    "Rating" : ['⭐ Invalid Rating / 5','Not Rated'],
    "Price" : ["Price Unavailable",None]
}

for key, value in dirty_pattern.items():
    df[key] = df[key].apply(lambda x: None if x in value else x)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df
```

```python
df.info()
```

```python
import re
# Remove the dollar sign and convert the 'Price' column to float
df['Price_in_dolar'] = df['Price'].apply(
    lambda x: float(re.search(r"[\d.]+", x).group()) if pd.notnull(x) else 0
)

# Convert 'Price_in_dolar' to rupiah
df['Price_in_rupiah'] = (df['Price_in_dolar'] * 16000).astype(float)

# Extract the numeric rating from the 'Rating' column and convert to float
df['Rating'] = df['Rating'].apply(lambda x: float(x.split()[1]))

# Extract the numeric value from the 'Colors' column and convert to float
df['Colors'] = df['Colors'].apply(lambda x: int(x.split()[0]))

# Convert 'Timestamp' to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df.drop(['Price'],inplace=True,axis=1)

df.info()
```
